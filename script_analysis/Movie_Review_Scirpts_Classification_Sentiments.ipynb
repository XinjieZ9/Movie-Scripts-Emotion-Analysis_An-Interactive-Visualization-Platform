```python
from transformers import pipeline
```

```python
summary_rows = []
target_emotions = ["sadness", "joy", "love", "anger", "fear", "surprise"]
```

```python
for csv_file in all_csv_files:
    movie_title = os.path.splitext(os.path.basename(csv_file))[0]  

    print(f"\nProcessing movie: {movie_title}")
    df = pd.read_csv(csv_file, encoding="utf-8")

  
    if "Sentence" not in df.columns:
        print(f"Warning: no 'Sentence' column found in {csv_file}, skipped.")
        continue

  
    emotion_sums = {emo: 0.0 for emo in target_emotions}
    total_lines = 0

  
    for text in df["Sentence"]:
        if not isinstance(text, str) or not text.strip():
            
            continue
        total_lines += 1

       
        results = classifier(
        text, 
        truncation=True,
        padding=True,
        max_length=512
    )
        # [[{'label': 'anger', 'score': 0.0003},
        #   {'label': 'fear', 'score': 0.0015},
        #   {'label': 'joy', 'score': 0.955},
        #   {'label': 'love', 'score': 0.004},
        #   {'label': 'sadness', 'score': 0.038},
        #   {'label': 'surprise', 'score': 0.0012}]]

   
        scores_dicts = results[0]  

        for d in scores_dicts:
            lab = d["label"].lower()
            sc = d["score"]
            if lab in emotion_sums:
                emotion_sums[lab] += sc

    if total_lines == 0:
        print(f"No valid lines found in {csv_file}, skipped.")
        continue


    emotion_avgs = {emo: (emotion_sums[emo] / total_lines) for emo in target_emotions}

    
    row_data = {"movie_title": movie_title}
    row_data.update(emotion_avgs)  
    summary_rows.append(row_data)

if len(summary_rows) == 0:
    print("No data to save. Maybe no valid CSV found?")
else:
    summary_df = pd.DataFrame(summary_rows)
 
    columns_order = ["movie_title"] + target_emotions
    summary_df = summary_df[columns_order]

    out_csv = "all_movies_subtitle_emotion_scores.csv"
    summary_df.to_csv(out_csv, index=False, encoding="utf-8")
    print(f"\nSaved final emotion scores to: {out_csv}")
```
